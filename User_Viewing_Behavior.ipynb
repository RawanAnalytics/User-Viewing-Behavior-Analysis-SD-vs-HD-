```python
# Import the required libraries

import pandas as pd
import pyxlsb
import numpy as np
import zipfile
```

```python
#After upload the data set and unzip the compressed file
with zipfile.ZipFile("User_Viewing_Behavior_Sample_data.zip", 'r') as zip_ref:
   zip_ref.extractall("temp_folder")
```

```python
#read the data set
file_path = "temp_folder/User_Viewing_Behavior_Sample_data.csv"
dataframe = pd.read_csv(file_path,encoding= 'latin1')
```

```python
# check the data shape
dataframe.shape
```

```python
# display the first 5 rows
dataframe.head()
```

```python
# Data Preprocessing on the input data
dataframe = dataframe.drop(columns=['Column1'])         # dropping the index column
dataframe['program_name'] = dataframe['program_name'].str.strip()  # trim spaces in movies names to avoid misspellings in input data
dataframe['date_'] = pd.to_datetime(dataframe['date_'])  # read date column as date data type
dataframe[['duration_seconds', 'season','episode','series_title','hd']] = dataframe[['duration_seconds', 'season','episode','series_title','hd']].apply(pd.to_numeric)  # read numeric columns as numeric data types
dataframe[['user_id_maped', 'program_name','program_class','program_desc','program_genre','original_name']] = dataframe[['user_id_maped', 'program_name','program_class','program_desc','program_genre','original_name']].astype(str) # read string columns as string data types
```

```python
# display the dataset after applying data types
dataframe.head()
```

```python
# describe the numeric values in the dataset
dataframe.describe()
```

```python
# check if any column has null value in the dataset
dataframe.isnull().any()
```

```python
#make a Copy of the data frame
df = dataframe.copy()
```

```python
# try to get the most watched movies (Total Views / Total Users Views / Total watch time)
# For series concatenated the Session episode to differentiate between episodes
grouped=df.copy()
grouped.loc[grouped['program_class'] == 'SERIES/EPISODES', 'program_name'] = grouped['program_name']+'_SE'+grouped['season'].astype(str)+'_EP'+grouped['episode'].astype(str)
grouped = grouped.groupby(['program_name','program_class'])\
.agg({'user_id_maped': [('co1', 'nunique'),('co2', 'count')],\
      'duration_seconds': [('co3', 'sum')] }).reset_index()
grouped.columns = ['program_name','program_class','No of Users who Watched', 'No of watches', 'Total watch time in seconds']
grouped['Total watch time in houres']=grouped['Total watch time in seconds']/3600
grouped = grouped.drop(columns=['Total watch time in seconds'])
grouped = grouped.sort_values(by=['Total watch time in houres', 'No of watches','No of Users who Watched'], ascending=False).reset_index(drop=True)

# show the result
grouped.head(35)
```

```python

# Use the same group by  before, but add complete background information about each program (hd,program_desc,program_genre).


grouped = df.copy()

grouped.loc[grouped['program_class'] == 'SERIES/EPISODES', 'program_name'] = (
    grouped['program_name'] + '_SE' + grouped['season'].astype(str) + '_EP' + grouped['episode'].astype(str)
)

grouped = grouped.groupby(['program_name','program_class']).agg({
    'user_id_maped': [('co1', 'nunique'), ('co2', 'count')],
    'duration_seconds': [('co3', 'sum')],
    'hd': [('hd', 'max')],
    'program_desc': [('program_desc', 'first')],
    'program_genre': [('program_genre', 'first')]
}).reset_index()

grouped.columns = [
    'program_name','program_class',
    'No of Users who Watched', 'No of watches',
    'Total watch time in seconds', 'hd',
    'program_desc', 'program_genre'
]

grouped['Total watch time in houres'] = grouped['Total watch time in seconds'] / 3600
grouped = grouped.drop(columns=['Total watch time in seconds'])

grouped = grouped.sort_values(
    by=['Total watch time in houres', 'No of watches','No of Users who Watched'],
    ascending=False
).reset_index(drop=True)

# show the result
grouped.head(35)
```

```python
#To download the results as Excel file
with pd.ExcelWriter('Sample_Results.xlsx') as writer:
      grouped.to_excel(writer, sheet_name='HD_NotHD_MOVIE_SERIES_EPISODES', index=False)
```
